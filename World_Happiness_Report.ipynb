```python
# Get top ten most generous countries in 2017,2018 and 2019
df2017_generous=df2017.sort_values('Generosity',ascending=False).head(10)
df2017_generous
```

```python
df2017_generous
```

```python
df2018_generous=df2018.sort_values('Generosity',ascending=False).head(10)
df2018_generous
```

```python
df2018_generous
```

```python
df2019_generous=df2019.sort_values('Generosity',ascending=False).head(10)
df2019_generous
```

```python
# Get data for generous countries common in all three years.
dfgenerous=df2017_generous.merge(df2018_generous,on="Country",how="inner")
dfgenerous
```

```python
dfgenerous2=dfgenerous.merge(df2019_generous,on="Country",how="inner")
dfgenerous2
```

```python
df3=dfgenerous2[['Country','Generosity_x','Generosity_y','Generosity']]
df3
df3.to_csv('Three_Years_Top_Generous.csv')
df3.to_excel('Three_Years_Top_Generous.xlsx')
```
