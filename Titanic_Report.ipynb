## Order of groupby changes the interpretation

```python
# Survival based on different classes
summsurvived=ttn.groupby(['pclass','survived','sex'])['pclass'].count()
summsurvived
```

```python
# Survival based on different sex
summsurvived=ttn.groupby(['sex','survived','pclass'])['pclass'].count()
summsurvived
```

a) More females in count survived ( females survived = 233 against male survived = 109). 
b) Percentage wise female survival was high (233/ 314 =74.2% against male (109/577 =18.89%)
c) Females who survived the class did not matter. Females who did not survive, most of them were from class 3
d) Males who survived the class did not matter. Males who did not survive, most of them were from class 3.
